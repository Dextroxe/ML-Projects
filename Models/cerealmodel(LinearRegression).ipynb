```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("cereal.csv")
data.head(5)
```

```python
data.shape
```

```python
data["name"].value_counts()
```

**we need to drop the first raw**

```python
data = data.drop(data.index[0])
data.head()
```

**name column is not necessery**

```python
data = data.drop("name",axis=1)
data.head()
```

```python
types = { "C" :1 , "H":0}
```

```python
data = data.replace({"type":types})
data.head()
```

```python
data1 = pd.get_dummies(data["mfr"])
data1.shape
```

```python
data2 = pd.concat([data,data1],axis='columns')
```

```python
data2 = data2.drop("mfr",axis=1)
```

## **Build the Model**

```python
X = data2.drop("rating",axis=1)
Y = data2["rating"]
```

```python
from sklearn.model_selection import  train_test_split

X_train, X_test , Y_train , Y_test =train_test_split(X,Y,test_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
def model_acc(model):
    model.fit(X_train,Y_train)
    acc = model.score(X_test,Y_test)
    print(str(model)+" : "+str(acc*100))
```

```python
LR = LinearRegression()
lasso = Lasso()
DT = DecisionTreeRegressor()
rf = RandomForestRegressor()

model_acc(LR)
model_acc(lasso)
model_acc(DT)
model_acc(rf)
```

**Linear regression has the best accuracy** </br> it has nearly 100% so no need to tune
